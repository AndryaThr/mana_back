```python
import sqlite3
```

```python
fihirana = open('000_FIHIRANA LIGUE.txt', 'r')
lines = fihirana.readlines()
```

```python
class Hira:

    def insert_into_db(self, database_path):
        self.con = sqlite3.connect(database_path)
        self.cur = self.con.cursor()
        self.cur.execute("""
            CREATE TABLE IF NOT EXISTS fihirana (
            id text primary key,
            lohateny varchar(100),
            laharana text(100),
            tononkira text);""")

        self.query = "insert into fihirana (id, laharana, tononkira, lohateny) values('"+ str(self.laharana)+"-"+str(self.lohateny) + "', "+ "'"+str(self.laharana)+"', '"+self.tononkira+"', '"+self.lohateny+"')"
        print(self.query)
        self.cur.execute(self.query)
        self.con.commit()
        self.con.close()

    def set_lohateny(self, lohateny):
        self.lohateny = lohateny.replace("'", "\"")

    def set_laharana(self, laharana):
        try: self.laharana = int(laharana)
        except: self.laharana = "".replace("'", "\"")

    def set_tononkira(self, tonokira):
        self.tononkira = tonokira.replace("'", "\"")

    def describe(self):
        print(self.lohateny)
        print(self.tononkira)
```

```python
tononkira = None
lohateny = None
i=-1

additive_title =[]
additive = False
pre_additive = False

for line in lines:
    if not additive:
        try:
            if not pre_additive:
                laharana = int(line)
            else:
                additive = True
            i += 1
            if tononkira is None:
                tononkira = ""
            else:
                hira = Hira()
                hira.set_laharana(i)
                hira.set_tononkira(tononkira.replace("\n", "<br/>"))
                hira.set_lohateny(lohateny.replace("\n", "<br/>"))
                hira.insert_into_db("baiboly.db")
                tononkira = ""
                lohateny = None

        except ValueError:
            if line == "ADDITIVE\n":
                pre_additive = True
            else:
                if lohateny is None:
                    lohateny = line
                tononkira += line

    else:
        if line.upper().replace("\n", "") == line.replace("\n", "") and line != "\n":
            i+=1
            if len(additive_title)>0:
                hira = Hira()
                hira.set_lohateny(additive_title.pop().replace("\n", "<br/>"))
                hira.set_tononkira(tononkira.replace("\n", "<br/>"))
                hira.set_laharana('None')
                hira.insert_into_db('baiboly.db')
            additive_title.append(line)
            tononkira = ""
        else:
            tononkira += line

        


print(i)
```
